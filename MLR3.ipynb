# HealthInsurence

```python
import pandas as pd
```

```python
df=pd.read_csv('../../datasets/Health_insurance.csv')
df
```

```python
df['sex_enc']=df['sex'].map({'female':1,'male':0})
df['smoker_enc']=df['smoker'].map({'yes':1,'no':0})
```

```python
df
```

```python
X=df[['age','sex_enc','bmi','children','smoker_enc']]
y=df['charges']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
age=int(input('Enter your age:'))
gender=input('Enter gender(male/female):')
gender= 0 if gender=='male' else 1
bmi=float(input('Enter bmi'))
children=int(input('Enter no of childrens'))
smoker=input('Do you smoke(yes/no)')
smoker= 1 if smoker=='yes' else 0

X_test=[[age,gender,bmi,children,smoker]]
print(model.predict(X_test)[0])
```
